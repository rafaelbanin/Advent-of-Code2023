```python
import re
list = ['one','two','three','four','five','six','seven','eight','nine']
cont = 0
a = open("Desafio1.txt","r")
for line in a.readlines():
    add = 1
    for word in list: #Code that transforms writen numbers into digits
        while (line.find(word) > -1):
            line = line[0:line.find(word)+1] + str(add) + line[line.find(word)+len(word)-1:len(line)] #1 and -1 for words that use the same letter "twone"
            print(line)
        add += 1
    n = re.findall(r'\d+', line) #finds all numbers in the line and put in list n
    n = "".join(n)
    res = [int(x) for a,x in enumerate(str(n))] 
    c =  res[0]*10 + res[-1]
    print(c)
    cont += c
a.close()
cont
```
